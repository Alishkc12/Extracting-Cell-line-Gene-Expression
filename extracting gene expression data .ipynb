```python
annotation_df = gene_df.copy()
```

```python
genes_df = annotation_df[annotation_df['type'] == 'gene']
```

```python
genes_df.attributes
```

```python
def parse_attributes(attr_str):
    attrs = {}
    for attr in attr_str.split(';'):
        if '=' in attr:
            key, value = attr.strip().split('=', 1)
            attrs[key] = value
    return attrs
```

```python
genes_df['attrs_dict'] = genes_df['attributes'].apply(parse_attributes)
```

```python
genes_df
```

```python
genes_df['gene_id'] = genes_df['attrs_dict'].apply(lambda x: x.get('gene_id'))
```

```python
genes_df['gene_type'] = genes_df['attrs_dict'].apply(lambda x: x.get('gene_type'))
```

```python
protein_coding_genes = genes_df[genes_df['gene_type'] == 'protein_coding']['gene_id'].tolist()
```
